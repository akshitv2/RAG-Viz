```python
import umap
import numpy as np
import matplotlib.pyplot as plt

# Assume 'data' is your array of shape (n_samples, 384)
data = np.random.rand(100, 384)  # Example data (replace with your own)

# UMAP for 2D
umap_2d = umap.UMAP(n_components=2, random_state=42)
embedding_2d = umap_2d.fit_transform(data)

# Plot 2D
plt.scatter(embedding_2d[:, 0], embedding_2d[:, 1])
plt.title("UMAP 2D Projection")
plt.show()

# UMAP for 3D
umap_3d = umap.UMAP(n_components=3, random_state=42)
embedding_3d = umap_3d.fit_transform(data)

# Plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(embedding_3d[:, 0], embedding_3d[:, 1], embedding_3d[:, 2])
plt.title("UMAP 3D Projection")
plt.show()
```
